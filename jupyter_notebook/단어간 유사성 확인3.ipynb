```python
import konlpy
from konlpy.tag import Okt


okt = Okt()
```

```python
columns_list = [] # dataframe 만들 리스트
label = [] # 질환명 넣어둘 라벨 리스트

for i in df.index:
  # (df['질환명'][i])
  for j in okt.nouns( df['증상'][i]):
    # if "." in j or "," in j:
    #   print(j)
    columns_list.append(j)
    label.append(df['질환명'][i])
```

```python
columns_list
```

```python
new_df = pd.DataFrame(data = None, columns=columns_list, index=columns_list)
new_df
```

```python
new_df['label'] = label
```

```python
new_df
```

```python
import time

aa = time.time()
cnt = 0
set_index = new_df.index


for i in new_df.index:
  for j in (new_df.columns)[:17087]: # label 컬럼은 제외
    if j in set_index and j == i:
      new_df[j][i] = 1
      cnt += 1

print(f'총 {cnt}회 연산')
print(time.time() - aa,"초")
# 대략 3억 회 연산?
# 2^17087 인건가

# 하... 이거 언제 다되냐...
```
