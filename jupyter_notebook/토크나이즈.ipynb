```python
# n-gram 어휘 벡터 생성
def getNgramWord(N,txt):
    txt = txt.split()
    ngrams = [txt[i:i+N] for i in range(len(txt)-N+1)]
    return ngrams
```

```python
# soynlp을 이용하여 학습이후, 응집도 계산

import urllib.request
from soynlp import DoublespaceLineCorpus
from soynlp.word import WordExtractor

urllib.request.urlretrieve("https://raw.githubusercontent.com/lovit/soynlp/master/tutorials/2016-10-20.txt", filename="2016-10-20.txt")

# 훈련 데이터를 다수의 문서로 분리
corpus = DoublespaceLineCorpus("2016-10-20.txt")

word_extractor = WordExtractor()
word_extractor.train(corpus)
word_score_table = word_extractor.extract()
```

```python
bigram = Phrases(word_lsts, min_count=1, threshold=1, delimiter="_")
bigram_phraser = Phraser(bigram)

trigram = Phrases(bigram_phraser[word_lsts], min_count=1, threshold=1.0, delimiter="_")
trigram_phraser = Phraser(trigram)

for w_l in word_lsts:
  bigram_s = bigram_phraser[w_l]
  trigram_s = trigram_phraser[w_l]
  print("유니그램 : {}".format(w_l))
  print("바이그램 : {}".format(bigram_s))
  print("트리그램 : {}".format(trigram_s))
  print("-"*40)
```

```python
import numpy as np
import random
import tensorflow as tf
from transformers import AutoTokenizer
from transformers import TFGPT2LMHeadModel
```

```python
model = TFGPT2LMHeadModel.from_pretrained('skt/kogpt2-base-v2', from_pt=True)
tokenizer = AutoTokenizer.from_pretrained('skt/kogpt2-base-v2')
```

```python
sent = '잠이 오노'
```

```python
input_ids = tokenizer.encode(sent)
input_ids = tf.convert_to_tensor([input_ids])
print(input_ids)
```

```python
output = model.generate(input_ids,
                        max_length=1000,
                        repetition_penalty=2.0,
                        use_cache=True)
output_ids = output.numpy().tolist()[0]
print(output_ids)
```

```python
tokenizer.decode(output_ids)
```

```python
output = model(input_ids)
top5 = tf.math.top_k(output.logits[0, -1], k=5)
```

```python
tokenizer.convert_ids_to_tokens(top5.indices.numpy())
```

```python
sent = '잠이 오노'
input_ids = tokenizer.encode(sent)

while len(input_ids) < 50:
    output = model(np.array([input_ids]))
    # Top 5의 단어들을 추출
    top5 = tf.math.top_k(output.logits[0, -1], k=5)
    # Top 5의 단어들 중 랜덤으로 다음 단어로 선택.
    token_id = random.choice(top5.indices.numpy())
    input_ids.append(token_id)

tokenizer.decode(input_ids)
```

```python
# https://wikidocs.net/159433

for sent in df['증상'].head(3):
  sent = str(sent)
  input_ids = tokenizer.encode(sent)
  input_ids = tf.convert_to_tensor([input_ids])
  # print(input_ids)

  output = model.generate(input_ids,
                        max_length=700,
                        repetition_penalty=2.0,
                        use_cache=True)
  output_ids = output.numpy().tolist()[0]
  # print(output_ids)
  tokenizer.decode(output_ids)

  output = model(input_ids)
  top5 = tf.math.top_k(output.logits[0, -1], k=5)

  tokenizer.convert_ids_to_tokens(top5.indices.numpy())

  input_ids = tokenizer.encode(sent)

  while len(input_ids) < 50:
    output = model(np.array([input_ids]))
    # Top 5의 단어들을 추출
    top5 = tf.math.top_k(output.logits[0, -1], k=5)
    # Top 5의 단어들 중 랜덤으로 다음 단어로 선택.
    token_id = random.choice(top5.indices.numpy())
    input_ids.append(token_id)

  print(tokenizer.decode(input_ids))
```
